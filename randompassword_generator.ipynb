```python
import random
import string

# --- 1. Define Character Sets ---
LOWERCASE = string.ascii_lowercase
UPPERCASE = string.ascii_uppercase
NUMBERS = string.digits
SYMBOLS = string.punctuation

# --- 2. Input and Validation Function ---
def get_user_preferences():
    """
    Collects and validates user input for password length and included character types.
    Returns a dictionary of validated preferences.
    """
    preferences = {}

    # Validate Password Length
    while True:
        try:
            length = int(input("Enter desired password length (minimum 4): "))
            if length < 4:
                print("âš ï¸ Password must be at least 4 characters long.")
                continue
            preferences['length'] = length
            break
        except ValueError:
            print("âŒ Invalid input. Please enter a whole number for the length.")

    # Helper function to get Yes/No input
    def get_yes_no_input(prompt):
        while True:
            response = input(f"{prompt} (y/n): ").lower().strip()
            if response in ['y', 'yes']:
                return True
            elif response in ['n', 'no']:
                return False
            else:
                print("âŒ Invalid input. Please enter 'y' or 'n'.")

    # Collect Character Type Preferences
    preferences['use_lowercase'] = True # Always true for this script
    preferences['use_uppercase'] = get_yes_no_input("Include Uppercase letters?")
    preferences['use_numbers'] = get_yes_no_input("Include Numbers?")
    preferences['use_symbols'] = get_yes_no_input("Include Symbols (e.g., !@#$)?")

    # Final Validation: Ensure at least one additional type is selected
    if not (preferences['use_uppercase'] or preferences['use_numbers'] or preferences['use_symbols']):
        print("\nðŸ”’ For a strong password, you must include at least one of Uppercase, Numbers, or Symbols.")
        # Recursively call the function if validation fails
        return get_user_preferences()

    return preferences

# --- 3. Core Generation Function ---
def generate_password(preferences):
    """
    Generates a strong, random password based on user preferences, guaranteeing
    at least one character from each requested type.
    """
    length = preferences['length']

    # 1. Initialize Lists and Pool
    character_pool = list(LOWERCASE)
    guaranteed_chars = []

    # 2. Build the Character Pool and Guaranteed List (guarantees one of each requested type)
    guaranteed_chars.append(random.choice(LOWERCASE))

    if preferences['use_uppercase']:
        character_pool.extend(UPPERCASE)
        guaranteed_chars.append(random.choice(UPPERCASE))

    if preferences['use_numbers']:
        character_pool.extend(NUMBERS)
        guaranteed_chars.append(random.choice(NUMBERS))

    if preferences['use_symbols']:
        character_pool.extend(SYMBOLS)
        guaranteed_chars.append(random.choice(SYMBOLS))

    # 3. Calculate and Fill the Remainder
    password_list = guaranteed_chars.copy()
    remaining_length = length - len(guaranteed_chars)

    # Fill the remaining space by drawing randomly from the full character_pool
    if remaining_length > 0:
        # Use random.choices for efficient selection of multiple characters
        remaining_chars = random.choices(character_pool, k=remaining_length)
        password_list.extend(remaining_chars)

    # 4. Shuffle and Output
    random.shuffle(password_list)
    return "".join(password_list)


# --- 4. Main Execution Block ---
if __name__ == "__main__":
    print("\n*** Random Password Generator ***")

    # 1. Get validated preferences from the user
    preferences = get_user_preferences()

    # 2. Generate the password
    try:
        password = generate_password(preferences)

        # 3. Print the result
        print("\n-------------------------------------------")
        print(f"âœ¨ Generated Password ({preferences['length']} chars): **{password}**")
        print("-------------------------------------------\n")

    except Exception as e:
        print(f"An unexpected error occurred during generation: {e}")
```
